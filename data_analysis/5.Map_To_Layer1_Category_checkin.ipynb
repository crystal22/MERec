```python
for index,data in data_city.iterrows():
    if data[5] in categoryTree['L2_Category_name'].tolist():
        #print("Yes2")
        a = categoryTree[categoryTree['L2_Category_name'].isin([data[5]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[5] in categoryTree['L3_Category_name'].tolist():
        #print("Yes3")
        a = categoryTree[categoryTree['L3_Category_name'].isin([data[5]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[5] in categoryTree['L4_Category_name'].tolist():
        #print("Yes4")
        a = categoryTree[categoryTree['L4_Category_name'].isin([data[5]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[5] in categoryTree['L5_Category_name'].tolist():
        #print("Yes5")
        a = categoryTree[categoryTree['L5_Category_name'].isin([data[5]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[5]=='Mall':
        data_city.loc[index,'L1_Category'] = 'Shop & Service'     
    elif data[5]=='Gas Station / Garage':  
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[5]=='Drugstore / Pharmacy':    
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[5]=='Light Rail':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[5]=='Spa / Massage':    
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[3]=='Volcanoes':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[5]=='Caf':    
        data_city.loc[index,'L1_Category'] = 'Drink'
    elif data[5]=='Ramen /  Noodle House':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[5]=='Gluten-free Restaurant':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[5]=='Caf':    
        data_city.loc[index,'L1_Category'] = 'Drink'
    elif data[5]=='Subway':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[5]=='Tennis':    
        data_city.loc[index,'L1_Category'] = 'Outdoors & Recreation'
    elif data[5]=='Ferry':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[5]=='Trails':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[5]=='Yogurt':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[5]=='Volcanoes':    
        data_city.loc[index,'L1_Category'] = 'Outdoors & Recreation'
    else:
        print(index,data[5])
    if index ==1000:
        print('1000')
    if index ==10000:
        print("10000")
    if index ==100000:
        print('100000')
    if index ==300000:
        print('300000')
print('Done')
```
