```python
import pandas as pd
import numpy as np       #载入必要的库
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city = 'Singapore'
data_city = pd.read_csv('../dataset/%s_new.csv'%(city), encoding='ISO-8859-1')
#data_city = pd.read_csv('../dataset/Calgary_new.csv', encoding='ISO-8859-1')
#data_city = pd.read_csv('../dataset/NewYork_new.csv', encoding='ISO-8859-1')
#data_city= pd.read_csv('../dataset/Singapore_new.csv', encoding='ISO-8859-1')
print(data_city.shape)
data_city
```

```python
categoryTree= pd.read_csv('../dataset/Foursquare_categories_split_Drink.csv', encoding='ISO-8859-1')
categoryTree
```

```python
for index,data in data_city.iterrows():
    if data[3] in categoryTree['L2_Category_name'].tolist():
        #print("Yes2")
        a = categoryTree[categoryTree['L2_Category_name'].isin([data[3]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[3] in categoryTree['L3_Category_name'].tolist():
        #print("Yes3")
        a = categoryTree[categoryTree['L3_Category_name'].isin([data[3]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[3] in categoryTree['L4_Category_name'].tolist():
        #print("Yes4")
        a = categoryTree[categoryTree['L4_Category_name'].isin([data[3]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[3] in categoryTree['L5_Category_name'].tolist():
        #print("Yes5")
        a = categoryTree[categoryTree['L5_Category_name'].isin([data[3]])]
        data_city.loc[index,'L1_Category'] = a['ï»¿L1_Category_name'].tolist()[0]
    elif data[3]=='Mall':
        data_city.loc[index,'L1_Category'] = 'Shop & Service'     
    elif data[3]=='Gas Station / Garage':  
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[3]=='Drugstore / Pharmacy':    
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[3]=='Light Rail':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[3]=='Spa / Massage':    
        data_city.loc[index,'L1_Category'] = 'Shop & Service'
    elif data[3]=='Volcanoes':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[3]=='Caf':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[3]=='Ramen /  Noodle House':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[3]=='Gluten-free Restaurant':    
        data_city.loc[index,'L1_Category'] = 'Food'
    elif data[3]=='Caf':    
        data_city.loc[index,'L1_Category'] = 'Drink'
    elif data[3]=='Subway':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[3]=='Tennis':    
        data_city.loc[index,'L1_Category'] = 'Outdoors & Recreation'
    elif data[3]=='Ferry':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[3]=='Trails':    
        data_city.loc[index,'L1_Category'] = 'Travel & Transport'
    elif data[3]=='Yogurt':    
        data_city.loc[index,'L1_Category'] = 'Food'
    else:
        print(index,data[3])
print('Done')
```
