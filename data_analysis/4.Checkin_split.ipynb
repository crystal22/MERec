```python
import pandas as pd
```

```python
Retrive_city = 'Singapore'
city_data = pd.read_csv('../dataset/%s_new.csv'%(Retrive_city), encoding='ISO-8859-1')
city_data
```

```python
for num in range(1,57):
    print("File %d Loading......"%(num))
    data_checkin = pd.read_table('../dataset/Foursq_fileParts/dataset_TIST2015_Checkins_%d.txt'%(num),sep='\t',header=None) 
    data_checkin.columns = ['UserId','VenueId','TimeStamp','TimeZone']
    data_checkin
    merge = pd.merge(data_checkin, city_data, how='inner', on=['VenueId'])
    print(merge.shape)
    #print(merge)
    if num ==1:
        new_data = merge
        print(new_data.shape)
    else:
        frames = [new_data, merge]
        new_data = pd.concat(frames)
        print(new_data.shape)
print("Finished!")
```

```python
new_data
```

```python
new_data = new_data.drop_duplicates()
new_data
```

```python
new_data = new_data.rename(columns={'Longitude':'Latitude ','Dimension':'Longitude '})
new_data
```

```python
new_data['Time'] = new_data['TimeStamp'].apply(lambda x: x.split()[3])
new_data['Time']
```

```python
mon_dic = {'Jan':'1', 'Feb':'2', 'Mar':'3', 'Apr':'4', 'May':'5', 'Jun':'6', 'Jul':'7', 'Aug':'8', 'Sep':'9', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
new_data['Date'] = new_data['TimeStamp'].apply(lambda x: x.split()[5] + mon_dic[x.split()[1]] + x.split()[2])
new_data['Date']
```

```python
new_data['Local_sg_time'] = new_data['Date'].map(str) + ' ' + new_data['Time']
new_data['Local_sg_time'] = new_data['Local_sg_time'].apply(lambda x: x.split()[0][0:4]+'/'+x.split()[0][4:-2]+'/'+x.split()[0][-2:] + ' ' + x.split()[1])
new_data['Local_sg_time'].apply(lambda x: x[:5]+'0'+x[5:] if len(x.split()[0].split('/')[1])==1 else x)
new_data['Local_sg_time']
```

```python
new_data = new_data.drop(columns=['TimeStamp'])
new_data
```

```python
new_data.to_csv("%s_checkin.csv"%(Retrive_city),index=False,sep=',')
print('%s Successful！'%(Retrive_city))
```
